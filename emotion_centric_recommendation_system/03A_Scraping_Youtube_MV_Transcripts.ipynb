### 02 Data Collection - Scraping Youtube

* Scrape English Transcript 100 "hottest" Music Videos from the 2020s as recommended by a playlist from Youtube

```python
def get_transcripts(video_ids):
    # Initialize an empty list to store the video IDs and transcripts
    data = []

    # Loop through each video ID in the list
    for video_id in video_ids:
        try:
            # Get the video transcript
            transcripts = YouTubeTranscriptApi.get_transcript(video_id)

            if transcripts:
                # Extract the transcripts as text
                subtitle_text = " ".join([entry["text"] for entry in transcripts])

                # Append the video ID and transcript to the data list
                data.append({'video_id': video_id, 'transcript': subtitle_text})

        except Exception as e:
            print(f"An error occurred with video {video_id}: {str(e)}")
            continue

    # Create a pandas DataFrame
    df = pd.DataFrame(data)

    return df
```

### 03 EDA & Cleaning

```python
# Display all data rows:
pd.set_option('display.max_rows', None)
pd.options.display.max_colwidth = 4000
lyrics
```

```python
# Convert strings in column 'transcript' to lowercase:
lyrics['transcript'] = lyrics['transcript'].str.lower()
```

```python
# Convert contraction words in column 'text':

# Define contraction function:
def expand_contractions(text):
    return contractions.fix(text)

# Apply contraction function:
lyrics['transcript'] = lyrics['transcript'].apply(expand_contractions)
```

```python
# Replace ♪ in column 'transcript' with ' ':
lyrics['transcript'] = lyrics['transcript'].str.replace(pat='♪', repl='.', regex=True)
```

```python
# Remove content within square brackets and the brackets themselves
lyrics['transcript'] = lyrics['transcript'].apply(lambda x: re.sub(r'\[[^\]]*\]', ' ', x))

# Remove content within parentheses and the parentheses themselves
lyrics['transcript'] = lyrics['transcript'].apply(lambda x: re.sub(r'\([^)]*\)', ' ', x))
```

```python
# Replace line of line break characters in column 'transcript' with ' ':
lyrics['transcript'] = lyrics['transcript'].str.replace(pat='\n', repl=' ', regex=True)
```

```python
# Remove leading spaces and punctuation from each transcript
lyrics['transcript'] = lyrics['transcript'].apply(lambda x: re.sub(r'^\W+', '', x))
```

```python
# Remove the space before a period or a question mark if it's preceded by a letter
lyrics['transcript'] = lyrics['transcript'].apply(lambda x: re.sub(r'([a-zA-Z]) \.', r'\1.', x))
lyrics['transcript'] = lyrics['transcript'].apply(lambda x: re.sub(r'([a-zA-Z]) \?', r'\1?', x))

# Remove the period or question mark if it's not preceded by a letter
lyrics['transcript'] = lyrics['transcript'].apply(lambda x: re.sub(r'\W\.', '', x))
lyrics['transcript'] = lyrics['transcript'].apply(lambda x: re.sub(r'\W\?', '', x))
```

```python
# Replace specified punctuations in column 'transcript' with ' ':
lyrics['transcript'] = lyrics['transcript'].str.replace(pat='[^a-zA-Z0-9\s.,?]', repl=' ', regex=True)
```

```python
# Replace numbers in column 'tweet' with ' ':
lyrics['transcript'] = lyrics['transcript'].str.replace(pat='[0-9]', repl=' ', regex=True)
```

```python
# Function to replace standalone single letters other than 'a' and 'i' with a space
def replace_standalone_letters(text):
    return re.sub(r'\b[^ai]\b', ' ', text, flags=re.IGNORECASE)

# Apply the function to the 'transcript' column
lyrics['transcript'] = lyrics['transcript'].apply(replace_standalone_letters)
```

```python
# Replace empty whitespaces in column 'transcript' with ' ':
lyrics['transcript'] = lyrics['transcript'].str.replace(pat='\s+', repl=' ', regex=True)
```

```python
# Remove trailing whitespaces in column 'transcript':
lyrics['transcript'] = lyrics['transcript'].str.strip()
```

### 04 Chunk down transcripts into length optimal for the Emotion-based Classification Model

```python
# Function to chunk transcripts in a DataFrame into segments with approximately 50 words each
# Input: DataFrame with 'video_id' and 'transcript' columns
# Output: DataFrame with 'video_id' and 'transcript' columns, containing segmented transcripts
def chunk_transcripts(df):
    new_rows = []
    for _, row in df.iterrows():
        video_id = row['video_id']
        transcript = row['transcript']
        sentences = sent_tokenize(transcript)
        chunks = []
        chunk = []
        word_count = 0
        for sentence in sentences:
            words = word_tokenize(sentence)
            if word_count + len(words) <= 50:
                chunk.append(sentence)
                word_count += len(words)
            else:
                if len(words) > 50:
                    # If a sentence has more than 50 words, split it into chunks
                    for word in words:
                        if word_count + 1 <= 50:
                            chunk.append(word)
                            word_count += 1
                        else:
                            chunks.append(' '.join(chunk))
                            chunk = [word]
                            word_count = 1
                else:
                    # If a sentence has less than 50 words, add it as a whole
                    chunks.append(' '.join(chunk))
                    chunk = [sentence]
                    word_count = len(words)
        chunks.append(' '.join(chunk))  # add the last chunk
        for chunk in chunks:
            new_rows.append({'video_id': video_id, 'transcript': chunk})
    return pd.DataFrame(new_rows)
```

```python
# Call function to chunk up lyrics Dataframe
chunked_lyrics = chunk_transcripts(lyrics)
```

```python
# Replace punctuations in column 'tweet' with ' ':
chunked_lyrics['transcript'] = chunked_lyrics['transcript'].str.replace(pat='[^a-zA-Z0-9\s]', repl=' ', regex=True)
```

```python
# Remove unecessary whitespaces:
chunked_lyrics['transcript'] = chunked_lyrics['transcript'].str.split(' ').str.join(' ')
```

```python
# Define a function that checks if a transcript has more than 50 words
def has_more_than_50_words(transcript):
    words = transcript.split()
    return len(words) > 50

# Apply this function to each transcript in the DataFrame
chunked_lyrics['more_than_50_words'] = chunked_lyrics['transcript'].apply(has_more_than_50_words)

# Filter the DataFrame to only include rows where 'more_than_50_words' is True
chunked_lyrics_filtered = chunked_lyrics[chunked_lyrics['more_than_50_words']]

# Print the filtered DataFrame
print(chunked_lyrics_filtered)
```

```python
# Drop the 'is_english' column from the DataFrame
chunked_lyrics = chunked_lyrics.drop(columns=['more_than_50_words'])
chunked_lyrics.head()
```
